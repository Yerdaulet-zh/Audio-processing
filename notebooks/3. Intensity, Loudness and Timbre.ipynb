<h3>Sound Power</h3>

<ul>
    <li>Sound power is the rate at which a sound source <b>emits acoustic energy.</b></li>
    <li>It describes how much sound energy is <b>produced per second, in all directions from the source.</b></li>
    <li>It is <b>measured in watts (W)</b>, just like electrical power.</li>
    <li>It is <b>independent of distance</b>—unlike sound pressure, which decreases as you move away.</li>
</ul>

If a speaker has a sound power of 1 watt, it means it is emitting 1 joule of sound energy every second, in every direction.

<hr>
<h3>Sound Power is Independent of Distance</h3>

This means that:
<ul>
    <li><b>The total energy</b> a sound source produces per second (sound power) <b>does not change</b>, no matter how far away you are from it.
    </li>
    <li>
        It’s a <b>fixed property</b> of the sound source itself.        
    </li>
</ul>

<hr>
<h3>Sound Pressure Depends on Distance</h3>

Sound Power is different from sound pressure, which is:

<ul>
    <li>What you <b>actually hear or feel</b> from sound (like volume).</li>
    <li>It <b>gets weaker</b> the farther you move from the source.</li>
    <li>This is why a loudspeaker sounds quieter when you walk away.</li>
</ul>


<h3>Analogy:</h3>

Think of a light bulb:
<ul>
    <li><b>Power</b> = how much light it gives off per second (always the same).</li>
    <li><b>Brightness you see</b> = how bright it looks to you (depends on how far you are).</li>
</ul>
<hr>

<h3>Sound Intensity</h3>

Definition:
Sound intensity is the amount of <b>sound power</b> that passes through a <b>specific area</b>.
It tells us <b>how concentrated</b> the sound energy is at a given spot.

Formula:

$$
I = \frac{P}{A}
$$
Where:

<ul>
    <li><b>I is intensity</b> (in watts per square meter, W/m²)</li>
    <li><b>P is sound power</b> (in watts, W)</li>
    <li><b>A is area</b> (in square meters, m²)</li>
</ul>

<hr>
<h3>How to Understand It</h3>

Imagine sound coming out of a speaker.

That sound energy spreads out over an area.

The farther you move from the source, the bigger the area becomes—and the <b>lower the intensity</b>, because the same amount of power is spread out more.

<hr>
<h3>Key Points:</h3>

<b>More concentrated power = higher intensity</b> (e.g., close to a loudspeaker).

<b>Spreads out over distance = lower intensity</b> (e.g., across a big hall).

It helps measure how much sound energy is hitting a surface, like your eardrum or a microphone.
<hr>

<h2>Threshold of Hearing</h2>

<ul>
    <li>Humans can perceive sounds with very small intensities.</li>
    <li>This lowest detectable sound intensity is called the <strong>Threshold of Hearing (TOH)</strong>.</li>
</ul>

$$
TOH = 10^{-12} \ \text{W/m}^2
$$

<h2>Threshold of Pain</h2>

<ul>
    <li>The upper limit of sound intensity that the human ear can tolerate without pain is called the <strong>Threshold of Pain (TOP)</strong>.</li>
    <li>Sounds at or above this level can cause physical pain and damage to hearing.</li>
</ul>

$$
TOP = 10 \ \text{W/m}^2
$$

<hr>
<h2>Sound Intensity Level</h2>

<ul>
    <li>Sound intensity level is measured on a <strong>logarithmic scale</strong>.</li>
    <li>It expresses the <strong>ratio</strong> between a given sound intensity and a reference intensity.</li>
    <li>The unit used is the <strong>decibel (dB)</strong>.</li>
    <li>The reference intensity is usually the <strong>Threshold of Hearing (TOH)</strong>:
        <br>
        \( I_0 = 10^{-12} \ \text{W/m}^2 \)
    </li>
</ul>

<h3>Formula:</h3>

$$
L = 10 \cdot \log_{10}\left( \frac{I}{I_0} \right) \ \text{dB}
$$

<ul>
    <li>\( L \): Sound intensity level (in dB)</li>
    <li>\( I \): Measured sound intensity (in W/m²)</li>
    <li>\( I_0 \): Reference intensity, usually \( 10^{-12} \ \text{W/m}^2 \)</li>
</ul>

<hr>
<h2>Why Do We Use a Logarithmic Scale for Sound Intensity?</h2>

<h3>1. Huge Range of Human Hearing</h3>

<p>
The human ear can detect sounds from:
</p>

<ul>
    <li>As quiet as:<br>
        <strong>
            \( I_0 = 10^{-12} \ \text{W/m}^2 \)
        </strong> &nbsp; (Threshold of Hearing)
    </li>
    <li>To as loud as:<br>
        <strong>
            \( I = 1 \ \text{W/m}^2 \)
        </strong> &nbsp; (near Threshold of Pain)
    </li>
</ul>

<p>
That’s a range of <strong>12 orders of magnitude</strong> — a trillion times difference!
</p>

<p><strong>A logarithmic scale compresses this enormous range</strong> into a more manageable scale from 0 dB to about 120 dB.</p>

<h3>2. Human Perception is Logarithmic</h3>

<ul>
    <li>Our ears don’t perceive sound intensity linearly.</li>
    <li>A sound that is <strong>10× more intense</strong> doesn’t sound 10× louder — it sounds just noticeably louder.</li>
    <li>A <strong>10 dB increase</strong> is perceived as <em>"twice as loud"</em> to the human ear.</li>
</ul>

<p><strong>So a logarithmic scale matches how we actually hear sound.</strong></p>

<hr>
<h2>"Every ~3 dB, intensity doubles"</h2>
<p><strong>The decibel formula is:</strong></p>

<p>
\( L = 10 \cdot \log_{10} \left( \frac{I}{I_0} \right) \)
</p>

<p><strong>Suppose you double the intensity:</strong></p>

<p>
\( \frac{I}{I_0} = 2 \)
</p>

<p><strong>Then:</strong></p>

<p>
\( L = 10 \cdot \log_{10}(2) \approx 10 \cdot 0.301 = 3.01 \ \text{dB} \)
</p>

<p>So, a <strong>3 dB increase</strong> means the sound intensity has <strong>doubled</strong> in physical energy.</p>

<h2>Loudness</h2>

<ul>
    <li><strong>Loudness</strong> is the <em>subjective perception</em> of how "loud" or "intense" a sound seems to a human listener.</li>
    <li>It is not the same as sound intensity (which is physical); loudness depends on how we <strong>experience</strong> the sound.</li>
    <li>Several factors influence loudness:
        <ul>
            <li><strong>Sound intensity</strong>: Louder sounds usually have higher intensity, but the relationship is not linear.</li>
            <li><strong>Frequency</strong>: We hear mid-range frequencies (like speech) as louder than very low or very high frequencies, even at the same intensity.</li>
            <li><strong>Duration</strong>: Very short sounds may seem quieter than sustained ones.</li>
            <li><strong>Age</strong>: Older people often lose sensitivity to higher frequencies, which affects perceived loudness.</li>
        </ul>
    </li>
    <li>Loudness is measured in <strong>phons</strong>, a unit that relates loudness to decibel levels at a reference frequency (usually 1,000 Hz).</li>
</ul>

<h3>Important:</h3>
<p>
While sound intensity level is measured in decibels (dB), loudness level is measured in phons, which takes human perception into account.
</p>

<h2>Timbre</h2>
<ul>
  <li><strong>Colour of sound</strong></li>
  <li>Difference between two sounds with same intensity, frequency, and duration</li>
  <li>Described with words like: <em>bright</em>, <em>dark</em>, <em>dull</em>, <em>harsh</em>, <em>warm</em></li>
</ul>
<hr>
<h2>Timbre is Subjective</h2>

<p>
While physical properties like <strong>frequency</strong> and <strong>intensity</strong> can be measured objectively,
<strong>timbre</strong> refers to how we perceive the <em>quality</em> or <em>character</em> of a sound — which is
influenced by many factors, such as:
</p>

<ul>
  <li>Harmonic content</li>
  <li>Attack and decay of the sound</li>
  <li>Presence of overtones or noise</li>
  <li>Listener’s own hearing sensitivity and experience</li>
</ul>

<p>
Two people might describe the same sound differently — one might call a violin "warm," another "harsh."
That’s why we use descriptive, <strong>subjective terms</strong> like <em>bright</em>, <em>dull</em>, <em>nasal</em>,
<em>smooth</em>, or <em>crisp</em> to talk about timbre.
</p>
<hr>
<h2>Features of Timbre</h2>

<ul>
  <li><strong>Timbre is multidimensional</strong> – It cannot be described by a single parameter like pitch or loudness. Timbre results from a complex combination of sound properties.</li>

  <li><strong>Sound Envelope</strong> – This refers to how a sound evolves over time, typically described by its <em>attack, decay, sustain,</em> and <em>release</em> (ADSR). The shape of this envelope influences how we perceive the sound.</li>

  <li><strong>Harmonic Content</strong> – The presence and relative strength of harmonics (overtones) above the fundamental frequency greatly affect a sound’s character. For example, a flute and a violin playing the same note have different harmonic structures, giving them distinct timbres.</li>

  <li><strong>Amplitude / Frequency Modulation</strong> – Variations in volume (amplitude modulation) or pitch (frequency modulation) over time can add richness or texture to a sound, influencing its perceived quality.</li>
</ul>

  <h1>Complex Sound</h1>

  <p>A <strong>complex sound</strong> is made up of simpler sinusoidal sounds (pure tones) added together.</p>

  <h2>Superposition of Sinusoids</h2>
  <p>A complex sound is the sum of several sine waves with different frequencies and amplitudes.</p>

  <div class="example">
    <strong>Example:</strong><br>
    A flute playing a note at 440 Hz doesn't just produce a pure tone. It includes:<br>
    - 440 Hz (fundamental frequency)<br>
    - 880 Hz (1st overtone, 2nd harmonic)<br>
    - 1320 Hz (2nd overtone, 3rd harmonic), etc.
  </div>

  <h2>Fourier Analysis</h2>
  <p>Using Fourier analysis, any complex sound can be broken down into its individual sinusoidal components.</p>

  <ul>
    <li>This is the basis of spectrograms and sound synthesis.</li>
    <li>Each component has a specific amplitude and frequency.</li>
  </ul>

  <h2>Characteristics</h2>
  <ul>
    <li><strong>Pitch:</strong> Determined by the fundamental frequency.</li>
    <li><strong>Loudness:</strong> Related to the total amplitude of all components.</li>
    <li><strong>Timbre:</strong> Depends on the combination and strength of the harmonics.</li>
  </ul>

  <h2>Summary</h2>
  <p>Complex sounds are everywhere in speech, music, and nature. Understanding them helps in audio processing, synthesis, and analysis.</p>

<h2>Fundamental Frequency</h2>
<p>
  The <strong>fundamental frequency</strong> is the <em>lowest</em> frequency of a sound. It determines the <strong>pitch</strong> you perceive.
</p>
<p>
  Example: If a note has a fundamental frequency of <code>440 Hz</code>, that’s the main pitch you hear.
</p>

<hr>

<h2>Harmonics</h2>
<p>
  <strong>Harmonics</strong> are <em>integer multiples</em> of the fundamental frequency. They contribute to the sound’s <strong>timbre</strong> (tone color).
</p>
<ul>
  <li>1st harmonic = f (fundamental)</li>
  <li>2nd harmonic = 2 × f</li>
  <li>3rd harmonic = 3 × f</li>
  <li>...</li>
</ul>

<p>For <code>440 Hz</code>:</p>
<ul>
  <li>1st harmonic = 440 Hz (fundamental)</li>
  <li>2nd harmonic = 880 Hz</li>
  <li>3rd harmonic = 1320 Hz</li>
  <li>4th harmonic = 1760 Hz</li>
</ul>

<hr>
