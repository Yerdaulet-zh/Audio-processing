# Understanding Audio: From Sound to Signal

---

## 1. What is Sound?

Sound is a **mechanical wave** that results from the **vibration of objects**. It travels through a medium (like air, water, or solid) and can be perceived by our ears.

### Key Concepts:

- **Sound is produced by vibration**
- **Vibrations cause air molecules to oscillate**
- **Oscillating air molecules create variations in pressure (sound waves)**

---

## 2. Sound as a Wave

Sound propagates in the form of **longitudinal waves** (not transverse like light).

### Important Properties:

| Property      | Description                                |
|---------------|--------------------------------------------|
| Frequency (Hz)| Number of oscillations per second (pitch)  |
| Amplitude     | Height of wave (volume/loudness)           |
| Wavelength    | Distance between peaks                     |
| Speed         | How fast it travels (depends on medium)    |

```python

````python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import (
    Audio, 
    YouTubeVideo,
    display, 
    HTML
)
````

````python
# Parameters
frequency = 20  # Hz
amplitude = 1
duration = 1  # seconds
sampling_rate = 16000  # Hz

t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
wave = amplitude * np.sin(2 * np.pi * frequency * t)

plt.plot(t, wave)
plt.title("Simple Sine Wave (Sound)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
````

````python
# Just play around with audio paramters and try to listen :)
Audio(wave, rate=sampling_rate)
````

<h2>What is a waveform?</h2>

A waveform is a <b>visual representation</b> of how a sound or signal changes over time.
<br>
In audio, it shows how the air pressure (or amplitude) goes up and down as the sound plays. It’s usually drawn as a wavy line on a graph, where:

<ul>
    <li>the horizontal axis is time</li>
    <li>the vertical axis is the signal’s strength (amplitude)</li>
</ul>

<b>It Carries multifactorial information:</b>
<ul>
    <li>
        <b>Frequency:</b> This refers to how often the waveform repeats per second and is measured in Hertz (Hz). 
        Higher frequencies correspond to higher-pitched sounds, like a whistle, while lower frequencies produce lower-pitched sounds, like a drum.
    </li>
    <li>
        <b>Intensity:</b> Also known as amplitude, this indicates how strong or loud the sound is. 
        On the waveform, higher peaks mean louder sounds, while smaller peaks represent quieter sounds.
    </li>
    <li>
        <b>Timbre:</b> This is the unique quality or tone of a sound that allows us to distinguish between different sound sources, even if they have the same pitch and loudness. 
        It’s shaped by the waveform’s complexity and the presence of overtones or harmonics.
    </li>
</ul>

<h2>Pitch</h2>
<img src="../images/pitch.png">
This Pitch-Frequency Chart visually explains how musical pitch (what we perceive as "high" or "low" notes) corresponds to physical frequency (in Hertz, Hz).
Key Concepts to Understand the Chart:

1. <b>Pitch (Vertical Axis)</b>

    Labeled as A₀, A₁, A₂, ..., A₆ – these are musical notes.

    Each "A" represents a musical note A in a different octave.

   For example:

        A₄ = 440 Hz, the standard tuning pitch (concert A).

        A₅ = 880 Hz, one octave higher.

        A₃ = 220 Hz, one octave lower.
<br>

2. <b>The Curve</b>

    Shows how pitch increases logarithmically as frequency increases exponentially.

    Each octave doubles the frequency:
   
    A0=27.5 Hz⇒A1=55 Hz⇒A2=110 Hz⇒A3=220 Hz⇒A4=440 Hz
   
    This doubling pattern gives pitch its <b>logarithmic</b> nature.


<b>Pitch is the way humans perceive frequency, and we perceive it on a logarithmic scale.</b>


This means:


To us, doubling the frequency sounds like the same “distance” in pitch (an octave).

    e.g., 220 Hz → 440 Hz → 880 Hz = same pitch step size (octaves)

The difference between 100 and 200 Hz feels bigger than the difference between 1000 and 1100 Hz, even though both are 100 Hz apart.

This is why pitch perception is <b>relative</b>, not absolute.

<img src="../images/midi_notes.png">

<b>MIDI (Musical Instrument Digital Interface)</b> notes are numbers from 0 to 127 that represent musical pitches in a standardized digital format.


1. Each MIDI note corresponds to a specific pitch (frequency in Hz)

    The MIDI number is just an index for a musical note.

    Higher number → Higher pitch

2. Middle C is MIDI note 60

    It’s also called C4

    Frequency ≈ 261.6 Hz

3. A4 = 440 Hz is MIDI note 69

    This is the standard tuning pitch used by orchestras.

4. One octave = 12 MIDI steps

    So:

        C4 = 60

        C5 = 72

        C3 = 48

    Each step (semitone) increases frequency by the 12th root of 2 <b>(≈ 1.0595)</b>

5. Formula to convert MIDI note to frequency:

$$
f = 2^{\frac{n - 69}{12}} \cdot 440
$$

        Where n is the MIDI number.

<h2>What Are Cents</h2>

A cent is a unit of measurement for musical pitch, just like degrees measure angles or milliseconds measure time.

It allows us to measure very small differences between two pitches, finer than semitones.


1. An octave = 1200 cents

    One octave means the frequency doubles.

    For example:

        A3 = 220 Hz

        A4 = 440 Hz → that’s one octave = 1200 cents

2. A semitone = 100 cents

    There are 12 semitones in an octave:


   12×100=1200 cents


   So, one semitone (e.g., C to C♯) = 100 cents


<h3>Why Use Cents?</h3>

Musicians and audio engineers often need to measure very fine pitch differences:

    For tuning instruments more precisely

    For comparing two close frequencies

    For measuring vibrato, glides, or microtonal music


<h3>Pitch Sensitivity in Humans</h3>

    10–25 cents is about the smallest pitch change that an average human can notice.

    Professional musicians or trained ears may detect changes as small as 5 cents.


<h3>Cents Formula</h3>

If you have two frequencies f1 and f2, the number of cents between them is:

<b>Cents=1200⋅log⁡2(f2 / f1)</b>

Example:

    f1=440 Hz,f2=450 Hzf1​=440 Hz,f2​=450 Hz

    Cents≈1200⋅log⁡2(450/440)≈39.1 centsCents≈1200⋅log2​(450/440)≈39.1 cents

So, the pitch of 450 Hz is 39 cents higher than 440 Hz.
