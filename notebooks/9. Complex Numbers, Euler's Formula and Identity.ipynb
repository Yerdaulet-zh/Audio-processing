##  1. What Are Complex Numbers?

A <b>complex number</b> is a number that combines a <b>real</b> part and an <b>imaginary</b> part.

<h4>General Form:</h4>

$$
z = a + bi
$$

<ul>
  <li><b>a</b> â†’ Real part: \( \text{Re}(z) = a \)</li>
  <li><b>b</b> â†’ Imaginary part: \( \text{Im}(z) = b \)</li>
  <li><b>i</b> â†’ Imaginary unit, where \( i^2 = -1 \)</li>
</ul>

<h4>Example:</h4>

$$
z = 3 + 4i \quad \text{(Real part: 3, Imaginary part: 4)}
$$

---

##  2. Understanding the Imaginary Unit \( i \)

The imaginary unit \( i \) is defined such that:

$$
i^2 = -1
$$

So we get:

$$
i^0 = 1,\quad i^1 = i,\quad i^2 = -1,\quad i^3 = -i,\quad i^4 = 1, \dots
$$

##  4. Operations with Complex Numbers

Let \( z_1 = a + bi \), \( z_2 = c + di \)

### âž• Addition:
$$
z_1 + z_2 = (a + c) + (b + d)i
$$

### âž– Subtraction:
$$
z_1 - z_2 = (a - c) + (b - d)i
$$

### âœ– Multiplication:
$$
z_1 z_2 = (ac - bd) + (ad + bc)i
$$

### âž— Division:
$$
\frac{z_1}{z_2} = \frac{a + bi}{c + di} = \frac{(a + bi)(c - di)}{c^2 + d^2}
$$

Multiply numerator and denominator by the <b>conjugate</b> of the denominator.

---

##  5. Complex Conjugate

The <b>conjugate</b> of \( z = a + bi \) is:

$$
\bar{z} = a - bi
$$

<b>Properties:</b>

- \( z \cdot \bar{z} = a^2 + b^2 = |z|^2 \)
- Useful in division and simplifying expressions

---

##  6. Modulus (Magnitude) of a Complex Number

The <b>modulus</b> or absolute value is the distance from the origin in the complex plane:

$$
|z| = \sqrt{a^2 + b^2}
$$

---

##  7. Argument (Phase Angle)

The <b>argument</b> is the angle (in radians) between the positive real axis and the line connecting the origin to the point:

$$
\arg(z) = \theta = \tan^{-1} \left( \frac{b}{a} \right)
$$

---

##  8. Polar and Exponential Form

A complex number can also be represented as:

### Polar Form:

$$
z = r(\cos \theta + i \sin \theta)
$$

Where:
- \( r = |z| \)
- \( \theta = \arg(z) \)

### Euler's Formula:

$$
z = re^{i\theta}
$$

---

##  9. De Moivreâ€™s Theorem

For integer \( n \), and \( z = re^{i\theta} \):

$$
z^n = r^n e^{in\theta} = r^n (\cos(n\theta) + i \sin(n\theta))
$$

---

##  10. Roots of Complex Numbers

To find \( n \)-th roots of \( z = re^{i\theta} \):

$$
\sqrt[n]{z} = r^{1/n} e^{i(\theta + 2k\pi)/n},\quad k = 0, 1, ..., n-1
$$

There are \( n \) distinct roots, equally spaced around a circle in the complex plane.

<h2>Understanding Complex Numbers in Polar Coordinates</h2>

<h3>Rectangular (Cartesian) Form</h3>

Any complex number <code>c</code> can be written as:

$$
c = a + ib
$$

<ul>
  <li><strong>a</strong> â†’ real part (x-axis)</li>
  <li><strong>b</strong> â†’ imaginary part (y-axis)</li>
  <li><strong>i</strong> â†’ imaginary unit where \( i^2 = -1 \)</li>
</ul>

---

<h3>Polar Form of a Complex Number</h3>

We can also represent a complex number in terms of <strong>magnitude</strong> and <strong>angle</strong>:

$$
c = |c| \cdot (\cos(\gamma) + i \sin(\gamma))
$$

Where:

<ul>
  <li><code>|c|</code> is the <strong>magnitude</strong> (or modulus) of the complex number</li>
  <li><code>Î³</code> is the <strong>angle</strong> (or argument) between the real axis and the vector to the point <code>c</code></li>
</ul>

---

<h3>Conversion Between Forms</h3>

Using the right triangle in the image:

<ul>
  <li><strong>Real part:</strong><br/>
  \( a = |c| \cdot \cos(\gamma) \)
  </li><br/>
  <li><strong>Imaginary part:</strong><br/>
  \( b = |c| \cdot \sin(\gamma) \)
  </li><br/>
  <li><strong>Magnitude (modulus):</strong><br/>
  \( |c| = \sqrt{a^2 + b^2} \)
  </li><br/>
  <li><strong>Angle (argument):</strong><br/>
  \( \gamma = \arctan\left(\frac{b}{a}\right) \)
  </li>
</ul>

---

<h3>Why Polar Form?</h3>

Polar form simplifies complex number operations:

<p><strong>Multiplication:</strong></p>

$$
z_1 \cdot z_2 = |z_1||z_2| \cdot [\cos(\theta_1 + \theta_2) + i \sin(\theta_1 + \theta_2)]
$$

<p><strong>Applications:</strong></p>

<ul>
  <li>Signal processing</li>
  <li>Fourier transforms</li>
  <li>Electrical engineering (phasors)</li>
  <li>Rotations in the complex plane</li>
</ul>

<h2>Additive Identity: <code>0</code></h2>

<h3>Definition:</h3>
<p>The <strong>additive identity</strong> is the number that, when added to any number, leaves the number unchanged.</p>

<h3>Why 0?</h3>
<p>Because:</p>
<ul>
  <li>For any number <code>a</code>, we have:</li>
  <li><code>a + 0 = 0 + a = a</code></li>
</ul>

<h4>Example:</h4>
<ul>
  <li><code>7 + 0 = 7</code></li>
  <li><code>(-5.3) + 0 = -5.3</code></li>
</ul>

<p>So "identity" here means: it doesnâ€™t change the value when added.</p>
