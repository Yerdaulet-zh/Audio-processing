### Intuition of the Fourier Transform in Sound

Real-world sounds are <b>complex</b>. They’re made up of <b>many frequencies happening at once, changing over time.</b> We hear this as the richness of music, speech, or environmental noise.


But here’s the powerful idea:
<ul>
    <li>
        <b>Any complex sound wave can be decomposed into a combination of simple sine and cosine waves of different frequencies, amplitudes, and phases.</b>
    </li>
</ul>

#### <span style="color:#007acc">This is the heart of the Fourier Transform (FT).</span>


<hr>

### Why Sine and Cosine?


Sine and cosine waves are the building blocks of <b>periodic signals</b>. They have:
<ul>
    <li>A single, <b>pure frequency</b></li>
    <li>A known <b>amplitude (loudness)</b></li>
    <li>A <b>phase (offset in time)</b></li>
</ul>

Even though <b>real sounds aren’t just clean sine waves,</b> the amazing result is:

#### <span style="color:#007acc">We can approximate any complex waveform by summing many sine and cosine waves at different frequencies.</span>

#### This is called a Fourier series for periodic signals and a Fourier transform for general (including non-periodic) signals.

<hr>

### What Does Fourier Transform Do?

Think of FT as a "frequency microscope":

You <b>give</b> it a <b>complex waveform</b> (e.g., a music note, a spoken word)

It tells you:
<ul>
    <li>What <b>frequencies are present</b> (like <b>440 Hz, 880 Hz, etc.</b>)</li>
    <li>How <b>strong each frequency</b> is (its <b>amplitude</b>)</li>
    <li><b>When it's happening (if using Short-Time FT)</b></li>
</ul>

You <b>go from</b> the <b>time domain</b> (waveform view) to the <b>frequency domain</b> (spectrum view).
<hr>

### Real Audio → Many Frequencies

<b>Real audio is not just one frequency</b>. For example:
<ul>
    <li>A <b>piano note contains</b> a <b>fundamental frequency<b/> (e.g., 440 Hz) <b>plus harmonics (multiples like 880 Hz, 1320 Hz…)</b></li>
    <li>A <b>spoken vowel</b> like <b>“ah” contains multiple resonant frequencies called formants</b></li>
</ul>

<b>Even noise</b> is just a <b>bunch of frequencies</b> added together randomly.

<h2 style="color:#007acc">Deeper Intuition Behind the Fourier Transform</h2>

<h3>The Key Idea:</h3>
<p>The Fourier Transform answers this question:</p>

<blockquote>
  “<strong>How much does this signal resemble a sinusoid of frequency <i>f</i>?</strong>”
</blockquote>

<p>To find out, it <strong>compares</strong> the signal to many sine and cosine waves (sinusoids) of different frequencies.</p>

<hr>

<h3>Step-by-Step Intuition:</h3>

<h4>1. Pick a frequency (e.g., 440 Hz)</h4>
<p>We ask: does the signal contain a 440 Hz tone?</p>

<h4>2. Multiply the signal by a sinusoid</h4>
<p>We take the signal <code>x(t)</code> and multiply it by a reference sinusoid, like <code>cos(2πft)</code> or <code>sin(2πft)</code>.</p>

<blockquote>
  If the signal contains this frequency, multiplying will <strong>reinforce</strong> it.  
  If not, the result will <strong>cancel out</strong> over time.
</blockquote>

<h4>3. Integrate or sum the result</h4>
<p>We add up the result over time:</p>

<p><b>Continuous case:</b></p>

$$
X(f) = \int_{-\infty}^{\infty} x(t) \cdot e^{-2\pi i f t} \, dt
$$

<p><b>Discrete (digital) case:</b></p>

$$
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-2\pi i k n / N}
$$

<p>The result is a complex number. Its:</p>
<ul>
  <li><strong>Magnitude</strong> tells us how strong this frequency is</li>
  <li><strong>Phase</strong> tells us how aligned or shifted it is</li>
</ul>

<hr>

<h3>High Magnitude = High Similarity</h3>

<ul>
  <li><b>Large magnitude</b>: signal matches the sinusoid — the frequency is present</li>
  <li><b>Small magnitude</b>: poor match — frequency is absent or very weak</li>
</ul>

<p>This is why the Fourier Transform can <strong>reveal hidden frequencies</strong> in a signal.</p>

<hr>

<h3>Example: A Pure Sine Wave at 440 Hz</h3>

<p>If our signal is:</p>

$$
x(t) = \sin(2\pi \cdot 440 t)
$$

<ul>
  <li>Compare to 440 Hz → ✅ perfect match → high magnitude</li>
  <li>Compare to 880 Hz, 220 Hz → ❌ poor match → near zero</li>
</ul>

<hr>

<h3>Repeat for All Frequencies</h3>
<p>We repeat this test for:</p>

<ul>
  <li>0 Hz (DC offset)</li>
  <li>100 Hz, 200 Hz, 300 Hz, …</li>
  <li>Up to half the sample rate (Nyquist limit)</li>
</ul>

<p>This gives a <strong>spectrum</strong> — a graph of frequency vs. magnitude.</p>

<hr>

<h3>Mental Picture</h3>

<p>Think of each frequency as a <strong>template</strong>, and we ask:</p>

<blockquote>
  “Does this frequency pattern appear in my signal?”
</blockquote>

<ul>
  <li>Yes → High response</li>
  <li>No → Low response</li>
</ul>

<hr>

<h3>Summary</h3>

<ul>
  <li>The Fourier Transform <strong>compares</strong> a signal with sinusoids of various frequencies.</li>
  <li>For each frequency, it computes a <strong>magnitude</strong> and <strong>phase</strong>.</li>
  <li>Frequencies with high magnitudes are <strong>present</strong> in the original signal.</li>
</ul>

<p>It’s like using a “frequency magnifying glass” to understand what your sound is made of.</p>

<hr>
<h2>Understanding the DFT Equation</h2>

$$
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-2\pi i k n / N}
$$

<ul>
  <li><code>x[n]</code>: time-domain input signal (e.g., audio samples)</li>
  <li><code>N</code>: total number of samples</li>
  <li><code>k</code>: frequency bin index (0 to N-1)</li>
  <li><code>X[k]</code>: complex value representing how much of the frequency <code>k/N</code> is present</li>
  <li><code>e^{-2πi k n / N}</code>: rotating complex sinusoid (both cosine and sine parts)</li>
</ul>

<p>Each <code>X[k]</code> gives us:</p>
<ul>
  <li><b>Magnitude</b>: how strong the <code>k</code>-th frequency is in the signal</li>
  <li><b>Phase</b>: where that frequency component starts in time</li>
</ul>

<p><b>Intuition:</b></p>
<blockquote>
  “We project the signal onto a rotating basis wave at frequency <code>k</code> and measure the similarity.”
</blockquote>

<p>By repeating this for all values of <code>k</code> from 0 to N-1, we get the full <b>frequency spectrum</b> of the signal.</p>

<h2>Reconstructing a Signal from Fourier Transform</h2>

<p>The Fourier Transform converts a signal from the <strong>time domain</strong> into the <strong>frequency domain</strong>.</p>

<p>But here's the magic:</p>

<blockquote>
  <strong>You can perfectly reconstruct the original signal by reversing the process!</strong>
</blockquote>

<hr>

<h3>How Reconstruction Works</h3>

<ul>
  <li><strong>Superimpose sinusoids</strong>: Rebuild the signal by adding many sine and cosine waves together</li>
  <li><strong>Weight them by magnitude</strong>: Use the amplitudes from the Fourier coefficients <code>|X[k]|</code></li>
  <li><strong>Use the phase</strong>: Shift each wave according to the phase angle of <code>X[k]</code></li>
</ul>

<p>This process is called the <strong>Inverse Discrete Fourier Transform (IDFT)</strong>:</p>

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] \cdot e^{2\pi i k n / N}
$$

<ul>
  <li><code>x[n]</code>: the reconstructed time-domain signal</li>
  <li><code>X[k]</code>: frequency domain representation (from DFT)</li>
  <li><code>N</code>: number of samples</li>
</ul>

<p>Each term in the sum is a sinusoid of frequency <code>k/N</code>, scaled by magnitude and shifted by phase.</p>

<hr>

<h3>Intuition</h3>

<p>The original signal is just a <strong>sum of sinusoids</strong>. The DFT tells us:</p>

<ul>
  <li>Which frequencies are present</li>
  <li>How strong they are (magnitude)</li>
  <li>Where they start (phase)</li>
</ul>

<p>The inverse DFT simply <strong>adds them back together</strong> to get the time-domain waveform.</p>

<hr>

<h3>Important Note</h3>

<blockquote>
  <strong>The original signal and its Fourier Transform contain the same information — just in different forms (time vs frequency).</strong>
</blockquote>

<p>No information is lost in the transform. It is fully invertible, assuming perfect precision.</p>

<hr>

<h3>Summary</h3>

<ul>
  <li>Fourier Transform: decompose signal into frequency components</li>
  <li>Inverse Transform: recombine those components to recreate the signal</li>
  <li>Every signal = sum of weighted, shifted sinusoids</li>
</ul>

<p>It’s like breaking a chord into notes (FT) and then playing those notes together to get the chord back (inverse FT).</p>


```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_path = "../audio/piano_c.wav"
ipd.Audio(audio_path)
```

```python
signal, sr = librosa.load(audio_path)
```

```python
plt.figure(figsize=(18, 8))
librosa.display.waveshow(signal, sr=sr, alpha=0.5)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# derive spectrum using FT
ft = sp.fft.fft(signal)

angle = np.angle(ft) # Phase in radians
magnitude = np.absolute(ft)

frequency = np.linspace(0, sr, len(magnitude)) 
```

```python
angle.shape, magnitude.shape
```

```python
ft[0], abs(ft[0])
```

```python
# plot spectrum
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], magnitude[:5000]) # magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
### Inverse FT
ifft_signal = sp.fft.ifft(ft)
ifft_signal = np.real(ifft_signal) # dropping the imaginary part
ipd.Audio(ifft_signal, rate=sr)
```

```python
# zomm in to the waveform
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], signal[10000:10400]) 
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# create a sinusoid

f = 523
phase = 0
phase2 = 0.5

sin = 0.5 * np.sin(2*np.pi * (f * t - phase))
sin2 = 0.5 * np.sin(2*np.pi * (f * t - phase2))

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], sin[10000:10400], color="r")
plt.plot(t[10000:10400], sin2[10000:10400], color="y")


plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# compare signal and sinusoids

f = 523
phase = 0.55

sin = 0.1 * np.sin(2*np.pi * (f * t - phase))

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], signal[10000:10400]) 
plt.plot(t[10000:10400], sin[10000:10400], color="r")

plt.fill_between(t[10000:10400], sin[10000:10400]*signal[10000:10400], color="y")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

This is a graphical interpretation of the inner product (dot product) between the signal and the sinusoid at 523 Hz.

If the signal and the sinusoid are in phase and match well → large yellow filled area → strong presence of that frequency.

If they don't match → the area will cancel out (positive × negative) → low or no presence of that frequency.
